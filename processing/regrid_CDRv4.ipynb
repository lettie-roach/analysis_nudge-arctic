```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import xesmf as xe
import os.path
```

CDRv4 doesn't have latitude and longitude, so grab them from CDRv3. Then regrid everything.

```python
wdir = '/glade/work/lettier/'
```

```python
cdrv4_ds = xr.open_dataset(wdir+'CDRv4/mon/nhdl/seaice_conc_monthly_nh_197901_n07_v04r00.nc').isel(tdim=0)
```

```python
cdrv3_ds = xr.open_dataset(wdir+'CDR/mon/seaice_conc_monthly_nh_1979-2018.nc').isel(time=0)
```

```python
cdrv4_ds['latitude'] = xr.DataArray(cdrv3_ds.latitude.values, dims = ('y','x'))
cdrv4_ds['longitude'] = xr.DataArray(cdrv3_ds.longitude.values, dims = ('y','x'))
```

```python
cesmgrid = xr.open_dataset(wdir+'/GRIDS/cesm1_tarea.nc')
```

```python
regridder = xe.Regridder(cdrv3_ds, cesmgrid, 'bilinear', filename='regrid_cdrv3_cesm.nc')
cdrv3_ds_regrid = regridder(cdrv3_ds)
cdrv3_ds_regrid

```

```python
regridder = xe.Regridder(cdrv4_ds, cesmgrid, 'bilinear', filename='regrid_cdrv4_cesm.nc')
cdrv4_ds_regrid = regridder(cdrv4_ds)
cdrv4_ds_regrid

```

```python
myfiles = sorted([f for f in os.listdir(wdir+'CDRv4/mon/nhdl/') if '.nc' in f])
```

```python
for f in myfiles:
    print(f)
    cdrv4_ds = xr.open_dataset(wdir+'CDRv4/mon/nhdl/'+f)
    cdrv4_ds['latitude'] = xr.DataArray(cdrv3_ds.latitude.values, dims = ('y','x'))
    cdrv4_ds['longitude'] = xr.DataArray(cdrv3_ds.longitude.values, dims = ('y','x'))
    
    regridder = xe.Regridder(cdrv4_ds, cesmgrid, 'bilinear',reuse_weights=True, filename='regrid_cdrv4_cesm.nc')
    cdrv4_ds_regrid = regridder(cdrv4_ds)
    cdrv4_ds_regrid.to_netcdf(wdir+'CDRv4/mon/nhdl/remap_cesmgrid/'+f)
    
```
